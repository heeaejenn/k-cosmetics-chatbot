# 3. WebBase Loader (웹 정보 크롤링)
- 웹 기반 문서 로더
- bs4 사용하여 웹 페이지 파싱 
- bs4.SoupStrainer 사용 -> 파싱할 요소 지정
- bs_kwargs 매개변수 통해 bs4.SoupStrainer의 추가 인수 지정 

## 3.1 페이지 한 개 크롤링

1. 뉴스 기사 크롤링 후 로드 

```python
import bs4
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader(
    web_path=('https://www.yna.co.kr/view/AKR20241005002500091?input=1195m'),
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(
            name=['h1', 'article'], # 들고 올 태그 여러 개 작성. -> 파싱할 요소 지정
            attrs={'class':['tit', 'story-news article']}
        )
    ),
    header_template={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36'} # header 필요시 넣는 곳
)

docs = loader.load()
print(type(docs))
print(len(docs))
print(type(docs[0]))
```

```python
print(docs[0].page_content)
```

    - False 처리하여 인증서를 검증하지 않도록 하여 에러 방지

# 4. Directory Loader
- 디렉토리 내의 문서 로드 
- 상대경로 
    - ./ (마침표 하나): 현재 작성중인 파일과 동일 폴더 기준
    - ../ (마침표 두 개): 현재 작성중인 파일의 상위 폴더 기준 
- glob: 로드할 파일의 형식을 지정할 수 있음 
    - *.확장자: 해당 폴더의 해당 확장자의 파일 
    - **/*.확장자: 해당 폴더의 하위까지 탐색
- use_multithreading
    - 보통 로딩은 하나의 스레드에서 이루어짐
    - 멀티 스레드를 활성화 함으로써 좀 더 빨리 많은 파일을 로드할 수 있음 
- loader_cls
    - 디렉토리에서 특정 확장자만 로드할 경우 로드해줄 로더 지정해주는 옵션 
    - import문에서 `ctrl + space`통해 로더를 찾아볼 수 있음 

```python
from langchain_community.document_loaders import DirectoryLoader, PyMuPDFLoader

loader = DirectoryLoader(
    # 디렉토리 경로 지정
    './',
    # 로드 할 확장자 지정
    glob='*.pdf', # <- @here
    # 멀티쓰레드 설정
    use_multithreading=True,
    # 특정 파일 로드 시, 로더 지정
    loader_cls=PyMuPDFLoader
)

docs = loader.load()
print(type(docs))
print(len(docs))

# 모든 pdf 파일 가져왔는데도 빠르게 처리됨.
# '**/*.pdf' 이렇게 하면 하위 폴더에 있는 pdf 파일까지 다 긁어옴. (@here)
```

# 5. UpstageLayoutAnalysisLoader
- 문서 분석 도구 
- 문서를 분석 해서 요소를 갖고옴 
- PDF, 이미지 등 다양한 형식에서 레이아웃을 분석해서 로드
- 문서에서 어느 부분이 제목인지, 본문인지, 표인지 등을 자동으로 인식하고 추출 
- OCR 기능 지원 (<-이미지로 텍스트 가져올 수 있음)
- Upstage AI에서 API키 발급 후 사용 가능
    - https://developers.upstage.ai/docs/getting-started/quick-start
    - UPSTAGE_API_KEY 변수명으로 .env에 api key 저장하기 
- 파라미터
    - file_path: 파일 경로
    - output_type: 출력 형식 (기본값: ['html', 'text])
    - split: 문서 분할 방식 ['None', 'element', 'page']
    - user_ocr = True (<-쓸지 말지 정할 수 있음)
    - exclude: 제외할 값 ['header', 'footer'] 

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain_upstage import UpstageLayoutAnalysisLoader

loader = UpstageLayoutAnalysisLoader(
    file_path='SW중심사회 9월호.pdf',
    output_type='text',
    split='page',
    use_ocr=True,
)

docs = loader.load()
print(type(docs))
print(len(docs))
```
